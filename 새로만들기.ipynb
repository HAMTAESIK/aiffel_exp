#데이콘 데이터 0,3,4,5번 항목 삭제

##aihub 데이터 카테고리명 변경 & 인덱스 제거 & 종합, 선거, 스포츠 +@ 제거

*   항목 추가
*   항목 추가



```python
condition3 = (aihub_data['topic_idx'] != '종합') & (aihub_data['topic_idx'] != '선거')
aihub_data = aihub_data[condition3]

aihub_data = aihub_data[aihub_data['topic_idx'] != '스포츠']
aihub_data = aihub_data[aihub_data['topic_idx'] != 'IT,과학']
aihub_data = aihub_data[aihub_data['topic_idx'] != '북한/한반도정세']
aihub_data = aihub_data[aihub_data['topic_idx'] != '부동산']
aihub_data = aihub_data[aihub_data['topic_idx'] != '기업']
```

```python
aihub_data['topic_idx'].value_counts()
```

##aihub 데이터 주제명 정수화

```python
#경제=1
#사회=2 
#보건의료환경=3
#정치=4
#교육입시=5

aihub_data.replace({'topic_idx': '경제'}, 1, inplace=True)
aihub_data.replace({'topic_idx': '사회'}, 2, inplace=True)
aihub_data.replace({'topic_idx': '보건/의료'}, 3, inplace=True)
aihub_data.replace({'topic_idx': '정치'}, 4, inplace=True)
aihub_data.replace({'topic_idx': '교육/입시/NIE'}, 5, inplace=True)
```

#dacon & aihub 데이터 합치기

```python
final = pd.concat([dacon_data,aihub_data] ,ignore_index=True)
```
