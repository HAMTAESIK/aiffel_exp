```python
def sampling_func(data, sample_pct):

    np.random.seed(123)

    N = len(data)

    sample_n = int(len(data)*sample_pct) # integer

    sample = data.take(np.random.permutation(N)[:sample_n])

    return sample
```

```python
sample_set = data1.groupby('topic_idx').apply(sampling_func, sample_pct=0.1)

sample_set.sort_index()
```

```python
sample_set = data1.groupby('topic_idx', group_keys=False).apply(sampling_func, sample_pct=0.1)

sample_set.sort_index()
```

```python
sample_set.columns = ['index','title','topic_idx']
```

```python
sample_set = sample_set.drop('index', axis=1)
```
